#  使用nba_api取得比賽數據

```python
from nba_api.stats.endpoints import TeamGameLog
from nba_api.stats.static import teams
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
```

# 找出球隊資料

```python
all_teams = teams.get_teams()
columns = [k for k in all_teams[0].keys()]
rows = []
for team in all_teams:
    rows.append([v for v in team.values()])
team_df = pd.DataFrame(data = rows, columns = columns)
team_df
```

# 建立球隊比賽數據dataframe

```python
columns = ['Team_ID', 'Game_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'W', 'L', 'W_PCT',
       'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA',
       'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']
rows = []
T_game = pd.DataFrame(data = rows, columns = columns)
T_game
```

# 利用得到的球隊id抓取其2013-2018球季的比賽數據¶

```python
seasons = []
for i in range(2013, 2019):
    seasons.append(i)
```

```python
for Tid in team_df["id"]:
    for season in seasons: 
        t_game_logs = TeamGameLog(team_id=Tid, season=season)
        gamelog = t_game_logs.get_data_frames()[0]
        T_game = pd.concat([T_game, gamelog], ignore_index=True, axis=0, sort=False)
```
