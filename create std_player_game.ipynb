```python
def create_sd_data(year):
    data_path = 'data/train/player_game_adv_' + year + '.csv'
    df = pd.read_csv(data_path)
#     print(df)
    columns = df.columns[4:]
    df_std = df.groupby(['GAME_ID', 'TEAM_ID'])[columns].std()
    del df_std['TEAM_ID']
#     print(df_std)
    std_path = 'data/train/std_player_game_adv_' + year + '.csv'
    df_std.to_csv(std_path)
```

```python
for year in range(2013, 2018):
    create_sd_data(str(year))
```
