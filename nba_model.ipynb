```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('data/train/train_final.csv', index_col=[0])
df_test = pd.read_csv('data/test/test_final.csv', index_col=[0])
```

```python
pd.set_option('display.max_columns', None)
```

```python
print("Train size:", len(df_train))
print("Test size:", len(df_test))
```

```python
df_test
```

## Handle categorical features

```python
# number of team
teams = np.unique(df_train['Team_ID'])
print(len(teams))
print(teams)
```

```python
df_train.groupby(['Team_ID', 'Match']).size()
```

```python
# convert team_id from int to string

df_train['Team_ID'] = df_train['Team_ID'].astype(str)
df_train['Match'] = df_train['Match'].astype(str)
df_test['Team_ID'] = df_test['Team_ID'].astype(str)
df_test['Match'] = df_test['Match'].astype(str)
```

```python
# Feature hashing scheme

from sklearn.feature_extraction import FeatureHasher
```

```python
fh = FeatureHasher(n_features=6, input_type='string')
hashed_teamid = fh.fit_transform(df_train['Team_ID'])
hashed_teamid = hashed_teamid.toarray()
df_x_train = pd.DataFrame(hashed_teamid, columns=['t0', 't1', 't2', 't3', 't4', 't5'])
```

```python
df_x_train
```

```python
hashed_match = fh.transform(df_train['Match'])
hashed_match = hashed_match.toarray()
```

```python
df_x_train = pd.concat([df_x_train, pd.DataFrame(hashed_match, columns=['m0', 'm1', 'm2', 'm3', 'm4', 'm5'])], axis=1)
```

```python
df_x_train
```

```python
hashed_teamid = fh.transform(df_test['Team_ID'])
hashed_teamid = hashed_teamid.toarray()

hashed_match = fh.transform(df_test['Match'])
hashed_match = hashed_match.toarray()
```

```python
df_x_test = pd.DataFrame(hashed_teamid, columns=['t0', 't1', 't2', 't3', 't4', 't5'])
df_x_test = pd.concat([df_x_test, pd.DataFrame(hashed_match, columns=['m0', 'm1', 'm2', 'm3', 'm4', 'm5'])], axis=1)
```

## Normalizing the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# select the features we would like to scale (all except Actual_PTS)

train_features = df_train.iloc[:, 2:-1]
```

```python
train_features
```

```python
# MinMaxScaler
scaler = MinMaxScaler()
train_features = scaler.fit_transform(train_features.values)
test_features = scaler.transform(df_test.iloc[:, 2:-1])
```

```python
train_features.shape
```

```python
train_features
```

```python
# concatenate categorical data with other features 

x_train = []

for i in range(len(df_train)):
    row = np.concatenate((df_x_train.values[i], train_features[i]))
    x_train.append(row)
    
x_test = []

for i in range(len(df_test)):
    row = np.concatenate((df_x_test.values[i], test_features[i]))
    x_test.append(row)
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
x_test = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
# prediction y is the Actual_PTS

y_train = df_train.iloc[:, -1]
y_test = df_test.iloc[:, -1]
```

## Build the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()
model.add(SimpleRNN(units=32, input_shape=(1,59), activation="relu"))
model.add(Dense(8, activation="relu")) 
model.add(Dense(1))
```

```python
model.compile(optimizer=RMSprop(), loss='mae')
```

```python
model.summary()
```

## Train the model

```python
history = model.fit(x_train, y_train, batch_size=16, epochs=10,
         validation_data=(x_test, y_test))
```

```python
pred = model.predict(x_test)
```

```python
# actual points and predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': pred.flatten()})
df
```

```python
# creating intervals, allowing prediction to be no more or less than 5 points

df["low"] = df["Predicted"] -5
df["high"] = df["Predicted"] +5
```
